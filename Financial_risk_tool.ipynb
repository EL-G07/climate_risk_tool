```python
data = pd.read_csv('climate_risk.csv')
data
```

```python
print(cs.w1)
```

```python
agri_w =cs.w1+cs.w2+cs.w3
cst_w=cs.w1+cs.w3+cs.w4
fld_w=cs.w1+cs.w3
en_w=cs.w2+cs.w1+cs.w3
total_w= cs.w1+cs.w2+cs.w3+cs.w4
```

```python
def risk_tier(row):
    y =(row.agri_loan_pct*agri_w+row.coastal_branches_pct*cst_w+row.flood_zone_pct*fld_w+row.energy_sector_pct*en_w)/(total_w*100)
    if y >= 0.66:
        return f"High"
    elif y< 0.66 and y>= 0.34:
        return f"Medium"
    elif y<0.34 and y>=0:
        return f"Low"

data['Climate_risk_tier']=data[['agri_loan_pct','coastal_branches_pct','flood_zone_pct','energy_sector_pct']].apply(risk_tier, axis=1)


```
